2. Buat Satu Buah Field Diskon (skor :15)

   Diskon Didapat Dari 10 Persen dari harga beli
   

```python
# Import library pandas
import pandas as pd

# Baca data (ganti 'data.csv' dengan nama file Anda)
databarang_dataset = pd.read_csv('databarang.csv')

# Buat kolom Diskon (10% dari Harga Beli)
databarang_dataset['Diskon'] = databarang_dataset['Harga Beli'] * 0.10

# Tampilkan hasil
print(databarang_dataset.head())

# Simpan kembali data ke file baru jika diperlukan
databarang_dataset.to_csv('databarang.csv', index=False)
```

3. Buat Satu Buah Field Pajak (Skor : 15)

   Pajak Didapat Dari 5 Persen dari Harga Beli
   
   

```python
# Import library pandas
import pandas as pd

# Baca data (ganti 'data.csv' dengan nama file Anda)
databarang_dataset = pd.read_csv('databarang.csv')

# Buat kolom Pajak (5% dari Harga Beli)
databarang_dataset['Pajak'] = databarang_dataset['Harga Beli'] * 0.05

# Tampilkan beberapa baris pertama untuk verifikasi
print(databarang_dataset.head())

# Simpan kembali data ke file baru (opsional)
databarang_dataset.to_csv('databarang.csv', index=False)
```

4. Buat Satu Buah Field Harga Jual (Skor : 15)

   Harga Jual Didapat Dari Harga Beli Dikurangkan Diskon Ditambah Pajak
   

```python
# Import library pandas
import pandas as pd

# Baca data (ganti 'data.csv' dengan nama file Anda)
databarang_dataset = pd.read_csv('databarang.csv')  # Sesuaikan dengan file data Anda

# Pastikan kolom Diskon dan Pajak sudah ada
# Hitung Harga Jual
databarang_dataset['Harga Jual'] = databarang_dataset['Harga Beli'] - databarang_dataset['Diskon'] + databarang_dataset['Pajak']

# Tampilkan beberapa baris pertama untuk verifikasi
print(databarang_dataset.head())

# Simpan kembali data ke file baru (opsional)
databarang_dataset.to_csv('databarang.csv', index=False)
```

5. Urutkan data terbesar ke terkecil berdasarkan harga jual (skor : 10)



```python
import pandas as pd

# Ganti 'data.csv' dengan nama file CSV Anda
databarang_dataset = pd.read_csv('databarang.csv')

# Mengurutkan data berdasarkan kolom "Harga Jual" (dari terbesar ke terkecil)
databarang_dataset = databarang_dataset.sort_values(by='Harga Jual', ascending=False)

# Menampilkan hasil
print(databarang_dataset)

# Jika ingin menyimpan hasil ke file baru
databarang_dataset.to_csv('databarang.csv', index=False)
```

6. Jika Harga Jual kurang dari 70000 Maka Status Audit Adalah Tidak, Diatas 70000 Status Audit Adalah Ya (skor : 20)



```python
import pandas as pd

# Ganti 'data.csv' dengan file CSV Anda
databarang_dataset = pd.read_csv('databarang.csv')

# Menambahkan atau memperbarui kolom 'Status Audit' berdasarkan kondisi
databarang_dataset['Status Audit'] = databarang_dataset['Harga Jual'].apply(lambda x: 'Tidak' if x < 70000 else 'Ya')

# Menampilkan hasil
print(databarang_dataset)

# Jika ingin menyimpan data ke file baru
databarang_dataset.to_csv('databarang.csv', index=False)
```

7. Buatlah layar keluaran visualisasi pie chart berdasarkan data file excel tingkat_kelulusan.csv (skor : 10)



```python
import pandas as pd
import matplotlib.pyplot as plt

# Membaca file
file_csv = 'tingkat_kelulusan.csv'
dataChart = pd.read_csv(file_csv)

# Menampilkan keseluruhan file untuk memahami struktur data
display(dataChart)

# Data untuk visualisasi
bulan = dataChart['Bulan']
nilai = dataChart['Tingkat Kelulusan']

# Urutan Warna
urutan_warna = [
    'gray', 'blue', 'pink', 'red', 'green', 'yellow', 
    'gray', 'blue', 'pink', 'red', 'green', 'yellow'
]

# Urutan bulan (mulai dari April)
urutan_bulan = bulan.tolist()[3:] + bulan.tolist()[:3]
urutan_nilai = nilai.tolist()[3:] + nilai.tolist()[:3]

# Membuat Pie Chart
plt.figure(figsize=(5, 5))
plt.pie(
    urutan_nilai, labels=urutan_bulan, autopct="%1.1f%%", 
    startangle=91, colors=urutan_warna, textprops={"fontsize":10}
)
plt.title("Distribusi Tingkat Kelulusan per Bulan", fontsize=14)
plt.show()
```

8. Buatlah layar keluaran visualisasi diagram garis berdasarkan data file excel tingkat_kelulusan.csv skor : 10)



```python
import pandas as pd
import matplotlib.pyplot as plt

# Membaca file
file_csv = "tingkat_kelulusan.csv"
dataChart = pd.read_csv(file_csv)

# Menampilkan keseluruhan file untuk memahami struktur data
display(dataChart)

# Membuat diagram garis berdasarkan data
plt.figure(figsize=(6.5, 5))
plt.plot(
    dataChart["Bulan"],
    dataChart["Tingkat Kelulusan"],
    linestyle="-",
    color="blue",
)

# Memberikan judul dan Label pada Sumbu X dan Y
plt.title("Data Mahasiswa Lulus", fontsize=14)
plt.xlabel("Bulan", fontsize=11)
plt.ylabel("Jumlah Mahasiswa Lulus", fontsize=11)
plt.xticks(rotation=90, fontsize=10)
plt.tight_layout()
plt.show()
```
